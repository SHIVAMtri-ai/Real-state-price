```python
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y.head()
```

```python
# train test dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train
```

```python
X_test
```

```python
# standardizing the dataset
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

# Model training

```python
from sklearn.linear_model import LinearRegression
```

```python
liner_regression = LinearRegression()
```

```python
liner_regression.fit(X_train, y_train)
```

```python
# print coefficient, and intercept
print(liner_regression.coef_)
```

```python
print(liner_regression.intercept_)
```

```python
# on which parameter the model is trained
liner_regression.get_params()
```

```python
# predition with the test data
lr_pred =liner_regression.predict(X_test)
```

```python
lr_pred
```

```python
# plot a sctter plot for a pred(comparing the x test pred data to ytest data)
plt.scatter(y_test, lr_pred)
```

```python
#calculate residuals
residual = y_test - lr_pred
```

```python
residual
```

```python
# plot the residual

sns.distplot(residual)
```

```python
#plot wrt pred and residual
plt.scatter(lr_pred, residual)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

print(mean_squared_error(y_test, lr_pred))
print(mean_absolute_error(y_test, lr_pred))

print(np.sqrt(mean_squared_error(y_test, lr_pred)))

```

# R^2 and adj R^2

```python
from sklearn.metrics import r2_score
score = r2_score(y_test, lr_pred)
```

```python
score
```

```python
#adj R^2 are have no specific library so we have write the manual formula of adj r squ
1- (1-score)* (len(y_test) - 1) / (len (y_test)-X_test.shape [1]-1)
```
